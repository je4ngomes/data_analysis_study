# Group by and Aggregation

```python
df = pd.read_csv('./Flavors.csv')
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.min()
```

```python
group_by_frame.max()
```

```python
group_by_frame.sum(numeric_only=True)
```

```python
group_by_frame.agg({
    'Flavor Rating': ['mean', 'max', 'count', 'sum'],
    'Texture Rating': ['mean', 'max', 'count', 'sum']
})
```

```python
df.groupby(['Base Flavor', 'Liked']) \
  .agg({
    'Flavor Rating': ['mean', 'max', 'count', 'sum'],
  })
```

```python
df.groupby('Base Flavor').describe()
```

# Merge, Join, and Concatenate

  - how='inner' (natural join)
  - how='outer' (full outer join)
  - how='left' (left outer join)
  - how='right' (right outer join)
  

```python
df1 = pd.read_csv('./LOTR.csv')
df2 = pd.read_csv('./LOTR 2.csv')
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_left', rsuffix='_right')
```
